# **Let's Open Your Virtual Screen**

To get started, you probably clicked a button like this <span style="background-color: green; color: white; padding: 2px 4px; border-radius: 4px;">Create codespace on master</span> to open your Codespace. Now, to see your turtle's drawings, you need to open a virtual screen.

Most of the time, this screen will open automatically. If it doesn't appear, look for the little monitor icon <img style="vertical-align:middle; border-radius: 8px; margin-bottom: 10px;" src="https://images.jointheleague.org/vscode/monitor_icon.png" height="25px" > in the top right corner of your editor and click on it.

You should now be looking at something like the image below. Click the *connect* button.
