We'll start by learning **turtle programming**, which lets you give commands to a virtual turtle, telling it where to move on the screen. It's like having a small robot artist you can control with code. To do this, we'll use a pre-installed Python tool called the **turtle module**.

You can tell your turtle to move forward, turn left or right, and change the color of the lines it draws. With these commands, you can create pictures like the one below.
