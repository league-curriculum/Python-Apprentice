### **How Tina Follows Commands**

Here is what some of the program will look like with <span title="Comments start with a '#' and explain what the code does" style="cursor: help;">comments</span> to explain what each line does:

```python
tina = turtle.Turtle()                  # Make a turtle named tina
tina.pencolor('blue')                   # Change tina's pen color to blue
tina.forward(200)                       # Move tina forward by 200 steps
tina.right(90)                          # Turn tina right by 90 degrees
```

- The code *before* the `#` is a **command** — it's an instruction that tells Tina exactly what to do, such as moving forward or changing color.
- The text *after* the `#` is a **comment** — it's a note for humans that explains what the command does. 

> **Tip:** Comments are ignored by Python and don't affect how the program runs.

## **Additional Help**

Reading documentation is one of the most useful skills a programmer can build. It helps you find commands, learn new features, and debug on your own.

### **How to Find Help**
* **Search Online:** Use Google or Bing to search for things like `python turtle commands` or `how to draw a circle with python turtle`.
* **Check Official Docs:** Look through the [Python Turtle documentation](https://docs.python.org/3/library/turtle.html). It lists every command you can use.
* **Look for Examples:** Websites like Stack Overflow or GeeksforGeeks often have code examples you can try.

### **Why Use Documentation?**

* **Learn Correct Syntax:** See exactly how to write a command so it works.
* **Discover Features:** find commands you haven't used yet.
* **Fix Errors:** Find out why your code might not be working and how to fix it.

> **Tip:** Copy example code and change things to see what happens. Experimentation is how most of this gets learned.
